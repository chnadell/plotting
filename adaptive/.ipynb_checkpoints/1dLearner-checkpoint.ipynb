```python
import adaptive
import random
import time
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

# define function to be sampled, with optional sleep to simulate a slow evaluation
a = 0.01
def func(x, wait=False):
    if wait:
        time.sleep(random.random()/10)
    return x + a**2/(a**2 + (x - .02)**2)

# define a learner which will suggest points to evaluate
learner = adaptive.Learner1D(func, bounds=(-1, 1))

# define a runner which will request points from the learner, then evaluate the function on those points
runner = adaptive.Runner(learner, goal=lambda l: l.loss() < 0.015)

# for live updates
# runner.live_info()
# runner.live_plot(updates_interval=0.1)
```

```python
# extract data from the learner
x = list(learner.data.keys())
y = list(learner.data.values())

# plot data to see the nonhomogeneous sampling
f = plt.figure(figsize = (16,8))
plt.scatter(x, y)
plt.show()
```
